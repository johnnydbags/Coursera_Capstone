CREATING THE DATAFRAME (PART 1)

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from IPython.display import Image 
from IPython.core.display import HTML 
from IPython.display import display_html
```

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

```python
import numpy as np
#clean the data
df1 = df[df.Borough != 'Not assigned']

# group by postal code
df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

# 'Not assigned' to names of Borough
df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])

df2
```

```python
df2.shape
```

ADDING LONGITUDE AND LATITUDE

```python
geo = pd.read_csv('https://cocl.us/Geospatial_data')
geo.head()
```

```python
df3 = pd.merge(df2,geo,on='Postal Code')
df3.head()
```

```python
df4 = df3[df3['Borough'].str.contains('Toronto',regex=False)]
df4
```

I'm going to go with the suggestion to work with only boroughs that contain the word Toronto

```python
import folium
torontomap = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(torontomap)
torontomap
```

```python
from sklearn.cluster import KMeans
k=5
torontocluster = df4.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(torontocluster)
kmeans.labels_
df4.insert(0, 'Cluster Labels', kmeans.labels_)
df4
```

```python
import matplotlib
from matplotlib import cm

clustermap = folium.Map(location=[43.651070,-79.347015],zoom_start=10)
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colorarray =  matplotlib.cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [matplotlib.colors.rgb2hex(i) for i in colorarray]
markercolors = []
for lat, lon, neighbourhood, cluster in zip(df4['Latitude'], df4['Longitude'], df4['Neighbourhood'], df4['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(clustermap)
       
clustermap
```
